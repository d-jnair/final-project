```python
import pandas as pd
import glob
import os

files = glob.glob("modis_2024_all_countries/modis/2024/*.csv")

dfs = []
for f in files:
    df = pd.read_csv(f)
    basename = os.path.splitext(os.path.basename(f))[0]
    parts = basename.split("_")
    country = " ".join(parts[2:])
    df["country"] = country
    dfs.append(df)

all_data = pd.concat(dfs, ignore_index=True)
all_data
```

```python
import pycountry_convert as pc
import pycountry

def country_to_continent(country_name):
    try:
        # get ISO alpha-2 from country name
        country = pycountry.countries.lookup(country_name)
        country_alpha2 = country.alpha_2
        
        # convert to continent code
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        
        # convert continent code to name
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    
    except:
        return None
```

```python
all_data["continent"] = all_data["country"].apply(country_to_continent)
all_data
```

```python
fixes = {
    "Cape Verde": "Africa",
    "Cote d Ivoire": "Africa",   # Côte d'Ivoire
    "Curacao": "North America",  # Caribbean → continent = North America
    "Democratic Republic of the Congo": "Africa",
    "Dem Rep Korea": "Asia",     # North Korea
    "Heard I and McDonald Islands": "Oceania",  # Officially part of the Antarctic region
    "Kosovo": "Europe",
    "Lao PDR": "Asia",           # Laos
    "Macedonia Former Yugoslav Republic of": "Europe",  # North Macedonia
    "Palestine": "Asia",
    "Republic of Congo": "Africa",   # Congo-Brazzaville
    "Republic of Korea": "Asia",     # South Korea
    "Reunion": "Africa",             # Admin region of France; geographically East Africa
    "Saint Helena": "Africa",
    "Swaziland": "Africa",           # Eswatini
    "The Gambia": "Africa",
    "Timor-Leste": "Asia",           # East Timor
    "Turkey": "Asia",                # Often Eurasia, but continent = Asia for most datasets
    "United States Minor Outlying Islands": "North America",
}

all_data["continent"] = all_data["continent"].fillna(all_data["country"].map(fixes))
all_data
```
